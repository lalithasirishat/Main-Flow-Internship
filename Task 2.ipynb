```python
# Step 3: Filter data based on conditions
filtered_df = df[df['ChipRate'] > 50] #Filter rows where 'ChipRate' > 50
print(filtered_df.head())
filtered_df_2 = df[df['BlowFlow'] < 100] #Filter rows where 'BlowFlow' < 100
print(filtered_df_2.head())
```

```python
# Step 4: Handle missing values
# Remove rows with missing values
df_cleaned = df.dropna()
print(df_cleaned.head())
# Fill missing values with 0
df_filled = df.fillna(0)
print(df_filled.head())
# Fill missing values in 'ChipRate' with the mean of 'ChipRate'
df['ChipRate'] = df['ChipRate'].fillna(df['ChipRate'].mean())
print(df.head())
```

```python
# Step 5: Calculate summary statistics
summary = df.describe()
print(summary)
# Specific statistics
mean_chiprate = df['ChipRate'].mean()
median_chiprate = df['ChipRate'].median()
std_chiprate = df['ChipRate'].std()
print(f"Mean ChipRate: {mean_chiprate}")
print(f"Median ChipRate: {median_chiprate}")
print(f"Standard Deviation of ChipRate: {std_chiprate}")
```
