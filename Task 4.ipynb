```python
import pandas as pd
# Extract the table from the image
data =pd.read_csv('USvideos.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.nunique() #Number of unique values in each column
```

```python
(df.isnull().sum()/(len(df)))*100  #Percentage of missing values in each column
```

```python
df_cleaned = df.dropna()
print("DataFrame after dropping rows with any missing values:")
print(df_cleaned.head())
# Fill missing values with 0 for the entire DataFrame
df_filled = df.fillna(0)
print("\nDataFrame after filling missing values with 0:")
print(df_filled.head())
# For demonstration purposes, let's assume the 'views' column has missing values.
# Fill missing values in 'views' with the mean of 'views'
df['views'] = df['views'].fillna(df['views'].mean())
print("\nDataFrame after filling missing values in 'views' with the mean of 'views':")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Distribution of views
plt.subplot(2, 2, 1)
sns.histplot(df['views'], bins=30, kde=True)
plt.title('Distribution of Views')

# Distribution of likes
plt.subplot(2, 2, 2)
sns.histplot(df['likes'], bins=30, kde=True)
plt.title('Distribution of Likes')

# Distribution of dislikes
plt.subplot(2, 2, 3)
sns.histplot(df['dislikes'], bins=30, kde=True)
plt.title('Distribution of Dislikes')

# Distribution of comment count
plt.subplot(2, 2, 4)
sns.histplot(df['comment_count'], bins=30, kde=True)
plt.title('Distribution of Comment Count')

plt.tight_layout()
plt.show()
```

```python
# Bar plot for the distribution of category_id
plt.figure(figsize=(10, 6))
sns.countplot(x='category_id', data=df, palette='viridis')
plt.title('Distribution of Category IDs')
plt.xlabel('Category ID')
plt.ylabel('Count')
plt.show()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

# Box plot for views
plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='views')
plt.title('Box Plot of Views')

# Box plot for likes
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='likes')
plt.title('Box Plot of Likes')

# Box plot for dislikes
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='dislikes')
plt.title('Box Plot of Dislikes')

# Box plot for comment count
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='comment_count')
plt.title('Box Plot of Comment Count')

plt.tight_layout()
plt.show()
```

```python
# Pairplot for numerical variables
sns.pairplot(df[['views', 'likes', 'dislikes', 'comment_count']])
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = df[['views', 'likes', 'dislikes', 'comment_count']].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Convert 'publish_time' to datetime
df['publish_time'] = pd.to_datetime(df['publish_time'])

# Extract date information
df['publish_date'] = df['publish_time'].dt.date

# Plot views over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='publish_date', y='views')
plt.title('Views Over Time')
plt.xlabel('Publish Date')
plt.ylabel('Views')
plt.show()
```

```python
# Box plots of numerical variables by category
plt.figure(figsize=(15, 10))

# Box plot for views by category
plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='category_id', y='views')
plt.title('Views by Category')

# Box plot for likes by category
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='category_id', y='likes')
plt.title('Likes by Category')

# Box plot for dislikes by category
plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='category_id', y='dislikes')
plt.title('Dislikes by Category')

# Box plot for comment count by category
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='category_id', y='comment_count')
plt.title('Comment Count by Category')

plt.tight_layout()
plt.show()
```
