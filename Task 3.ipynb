```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the image
data =pd.read_csv('householdtask3.csv')

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(data)

# Bar chart
fig, ax1 = plt.subplots(figsize=(14, 7))
bar_width = 0.25
index = np.arange(len(df))

bar1 = ax1.bar(index, df['tot_hhs'], bar_width, label='tot_hhs')
bar2 = ax1.bar(index + bar_width, df['own'], bar_width, label='own')
bar3 = ax1.bar(index + 2 * bar_width, df['own_wm'], bar_width, label='own_wm')

ax1.set_xlabel('Index')
ax1.set_ylabel('Number of Households')
ax1.set_title('Households Overview')
ax1.set_xticks(index + bar_width)
ax1.set_xticklabels(index)
ax1.legend()

# Line chart 
fig, ax2 = plt.subplots(figsize=(14, 7))


ax2.plot(df.index, df['age'], label='Age', marker='o', color='blue')
ax2.set_xlabel('Index')
ax2.set_ylabel('Age', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

ax3 = ax2.twinx()
ax3.plot(df.index, df['size'], label='Size', marker='o', color='green')
ax3.plot(df.index, df['income'], label='Income', marker='o', color='orange')
ax3.set_ylabel('Size / Income', color='green')
ax3.tick_params(axis='y', labelcolor='green')

fig.tight_layout()
ax2.set_title('Age, Size, and Income')

# Combine legends
lines, labels = ax2.get_legend_handles_labels()
lines2, labels2 = ax3.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```
